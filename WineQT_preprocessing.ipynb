```python
import numpy as np
import pandas as pd
```

```python
raw_data = pd.read_csv('drive/MyDrive/WineQT/WineQT.csv')
raw_data.head()
```

```python
first_data = raw_data.copy()
first_data.describe()
```

```python
second_data = first_data.drop(['Id'],axis=1)
second_data
```

```python
second_data.isnull()
```

```python
second_data.isnull().sum()
```

>> Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(second_data.drop(['quality'],axis=1))
```

```python
scaled_inputs = scaler.transform(second_data.drop(['quality'],axis=1))
```

```python
second_data.columns.values
```

```python
scaled_inputs = pd.DataFrame(data = scaled_inputs,columns = ['fixed acidity', 'volatile acidity', 'citric acid',
       'residual sugar', 'chlorides', 'free sulfur dioxide',
       'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol',])
```

```python
scaled_inputs.describe()
```

```python
targets = second_data['quality']
```

```python
targets_cat = targets.map({0:'poor',1:'poor',2:'poor',3:'poor',4:'poor',5:'poor'
                           ,6:'avg',
                           7:'good',8:'good',9:'good',10:'good'})
```

```python
targets_dummies = pd.get_dummies(targets_cat)
```

```python
targets_dummies = targets_dummies[['poor','avg','good']]
```

```python
targets_dummies
```

```python
targets_dummies['poor'].sum(),targets_dummies['avg'].sum(),targets_dummies['good'].sum()
```

>> Saving scaler object inside file using Pickle

```python
import pickle
with open('drive/MyDrive/WineQT/WineQT_Scaler','wb') as file:
  pickle.dump(scaler,file)
```

>> Splitting data into train(85%), test(10%), val(5%)

```python
from sklearn.model_selection import train_test_split
train_inputs,test_and_val_inputs,train_targets,test_and_val_targets = train_test_split(scaled_inputs,targets_dummies,test_size=0.15,random_state=422,shuffle=True)
```

```python
train_inputs.describe()
```

```python
test_and_val_targets.shape,test_and_val_inputs.shape
```

```python
test_inputs,val_inputs,test_targets,val_targets = train_test_split(test_and_val_inputs,test_and_val_targets,
                                                                     test_size=0.33,shuffle=True,random_state=32)
```

```python
test_inputs.shape,test_targets.shape,val_inputs.shape,val_targets.shape
```

>> Deploying into NPZ files

```python
np.savez('drive/MyDrive/WineQT/WineQT_train_data',inputs=train_inputs,targets=train_targets)
```

```python
np.savez('drive/MyDrive/WineQT/WineQT_test_data',inputs=test_inputs,targets=test_targets)
```

```python
np.savez('drive/MyDrive/WineQT/WineQT_val_data',inputs=val_inputs,targets=val_targets)
```
